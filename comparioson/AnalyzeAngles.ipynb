```python
from statistics import median
from statistics import stdev

from matplotlib import pyplot as plt
import matplotlib as mpl
```

```python
f_sm = open('debug_output_small.txt')
f_lg = open('debug_output_big.txt')
f_ha = open('debug_output_haar.txt')

sm = f_sm.read().split()
lg = f_lg.read().split()
ha = f_ha.read().split()

sm = [float(i) for i in sm]
lg = [float(i) for i in lg]
ha = [float(i) for i in ha]

print(len(sm), len(lg), len(ha))

print(sm[:10])
print(type(sm))
```

```python
sm = [i for i in sm if i > 0]
lg = [i for i in lg if i > 0]
ha = [i for i in ha if i > 0]

print(len(sm), len(lg), len(ha))
```

### Среднее

```python
av_sm = sum(sm)/len(sm)
av_lg = sum(lg)/len(lg)
av_ha = sum(ha)/len(ha)

print(av_sm, av_lg, av_ha)
```

### Медиана

```python
m_sm = median(sm)
m_lg = median(lg)
m_ha = median(ha)

print(m_sm, m_lg, m_ha)
```

### Стандартное отклонение

```python
d_sm = stdev(sm)
d_lg = stdev(lg)
d_ha = stdev(ha)

print(d_sm, d_lg, d_ha)
```

```python
sm_x = list(range(len(sm)))
lg_x = list(range(len(lg)))
ha_x = list(range(len(ha)))

print(len(sm_x))

#mpl.rc('xtick', labelsize=20) 
mpl.rcParams.update({'font.size': 22})
plt.figure(figsize=(15,10), dpi=300)

plt.scatter(sm_x, sm, label='small')
plt.scatter(lg_x, lg, label='large')
plt.scatter(ha_x, ha, label='haar')
plt.legend()
plt.ylabel('Значение угла, градусы')
plt.xlabel('№ измерения')
plt.show()
```
